```python
import numpy as np
import pandas as pd
import os
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
directory = './datasets/twitters/'
files = [file for file in os.listdir(directory) if file.endswith('.csv')]

counter = 0
stocks = {}
for file in files:
    df = pd.read_csv(directory + file)
    df['date'] = pd.to_datetime(df['date']).dt.tz_convert('US/Eastern')
    df.set_index('date', inplace=True)
    df.index = pd.DatetimeIndex(df.index).to_period('ms')
    df.sort_index(inplace=True)
    
    sent = SentimentIntensityAnalyzer()
    df['vader_sentiment'] = df['tweet_text'].apply(lambda x: sent.polarity_scores(str(x))['compound'])
    
    stocks[file[:-11]] = df
    df.to_csv(f'./datasets/twitters/withsent/{df["company"][0]}.csv', index=True)
    counter+=1
    print(f'exported {df["company"][0]}.  {counter}/{len(files)}')
    
```
